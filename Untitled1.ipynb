```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger une image depuis l'ordinateur
uploaded = files.upload()

# Récupérer le nom de l'image uploadée
image_name = list(uploaded.keys())[0]

# Charger l'image en niveaux de gris
image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Erreur : Impossible de charger l'image.")
else:
    print("Image chargée avec succès !")
```

```python
# Affichage de l'image
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title('Image Originale')

# Calcul et affichage de l'histogramme
histogram, _ = np.histogram(image.flatten(), 256, [0,256])

plt.subplot(1,2,2)
plt.plot(histogram, color='black')
plt.title('Histogramme')
plt.xlabel('Niveau de gris')
plt.ylabel('Nombre de pixels')

plt.show()
```

```python
# Clipping (transformation linéaire)
image_clipped = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)

# Égalisation d’histogramme
image_equalized = cv2.equalizeHist(image)

# Affichage des résultats
plt.figure(figsize=(10,6))

plt.subplot(2,3,1)
plt.imshow(image, cmap='gray')
plt.title('Image Originale')

plt.subplot(2,3,2)
plt.imshow(image_clipped, cmap='gray')
plt.title('Image après Clipping')

plt.subplot(2,3,3)
plt.imshow(image_equalized, cmap='gray')
plt.title('Image après Égalisation')

plt.subplot(2,3,4)
histogram_original, _ = np.histogram(image.flatten(), 256, [0,256])
plt.plot(histogram_original, color='black')
plt.title('Histogramme Original')

plt.subplot(2,3,5)
histogram_clipped, _ = np.histogram(image_clipped.flatten(), 256, [0,256])
plt.plot(histogram_clipped, color='red')
plt.title('Histogramme Clipping')

plt.subplot(2,3,6)
histogram_equalized, _ = np.histogram(image_equalized.flatten(), 256, [0,256])
plt.plot(histogram_equalized, color='blue')
plt.title('Histogramme Égalisé')

plt.tight_layout()
plt.show()
```
